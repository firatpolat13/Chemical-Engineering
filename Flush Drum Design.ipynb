```python
T = 298 #K
T_ref = 273
T_feed = 383
P = 2 #bar
VF_EST = 0.5

for i in range(10):
    
    CO = {"A" : 3.365, "B" : 230.272, "C" : -13.15, "Fi" : 19.040}
    CO2 = {"A" : 6.939, "B" : 1347.79, "C" : -0.16, "Fi" : 48.137}
    H2 = {"A" : 3.046, "B" : 71.6152, "C" : 3.19, "Fi" : 58.084}
    CH4 = {"A" : 3.737, "B" : 229.639, "C" : -7.16, "Fi" : 17.892}
    H2O = {"A" : 5.074, "B" : 1657.46, "C" : -46.13, "Fi" : 13.771}
    
    CO_Cp = 29.651 + (-0.0007 * T_feed) + (0.0000813 * T_feed**2) + ((-0.094 * 10**-7) * T_feed**3) + ((0.108*10**-11)* T_feed**4)
    CO2_Cp = 29.268 + (-0.0224 * T_feed) + (0.0002653 * T_feed**2) + ((-4.153*10**-7) * T_feed**3) + ((20.057*10**-11)* T_feed**4)
    H2_Cp = 27.004 + (0.0119 * T_feed) + (-0.0000241 * T_feed**2) + ((0.215*10**-7) * T_feed**3) + ((-0.615*10**-11)* T_feed**4)
    CH4_Cp = 36.155 + (-0.0511 * T_feed) + (0.0002215 * T_feed**2) + ((-0.094*10**-7) * T_feed**3) + ((0.108*10**-11)* T_feed**4)
    H2O_Cp = 33.763 + (-0.0006 * T_feed) + (0.0000224 * T_feed**2) + ((-0.1*10**-7) * T_feed**3) + ((0.11*10**-11)* T_feed**4)
    H2O_Liq_Cp = -203.606 + (1523.29 * T/1000) + (-3196.413 * (T/1000)**2) + (2474.46 * (T/1000)**3) + (3.856/(T/1000)**2)
    
    CO_CpV = 29.651 + (-0.0007 * T) + (0.0000813 * T**2) + ((-0.094 * 10**-7) * T**3) + ((0.108*10**-11)* T**4)
    CO2_CpV = 29.268 + (-0.0224 * T) + (0.0002653 * T**2) + ((-4.153*10**-7) * T**3) + ((20.057*10**-11)* T**4)
    H2_CpV = 27.004 + (0.0119 * T) + (-0.0000241 * T**2) + ((0.215*10**-7) * T**3) + ((-0.615*10**-11)* T**4)
    CH4_CpV = 36.155 + (-0.0511 * T) + (0.0002215 * T**2) + ((-0.094*10**-7) * T**3) + ((0.108*10**-11)* T**4)
    H2O_CpV = 33.763 + (-0.0006 * T) + (0.0000224 * T**2) + ((-0.1*10**-7) * T**3) + ((0.11*10**-11)* T**4)
    
    CO_HCond = -6450.7
    CO2_HCond = -25246
    H2_HCond = -904.35
    CH4_HCond = -8190

    
    Ft = CO["Fi"] + CO2["Fi"] + H2["Fi"] + CH4["Fi"] + H2O["Fi"]
    V = Ft * VF_EST 
    L = Ft - V

    CO_fraction = CO["Fi"] / Ft
    CO2_fraction = CO2["Fi"] / Ft
    H2_fraction = H2["Fi"] / Ft
    CH4_fraction = CH4["Fi"] / Ft
    H2O_fraction =  H2O["Fi"] / Ft
    
    CO_Hf = CO_fraction * CO_Cp * (T_feed - T_ref)
    CO2_Hf = CO2_fraction * CO2_Cp * (T_feed - T_ref)
    H2_Hf = H2_fraction * H2_Cp * (T_feed - T_ref)
    CH4_Hf = CH4_fraction * CH4_Cp * (T_feed - T_ref)
    H2O_Hf = H2O_fraction * H2O_Cp * (T_feed - T_ref)
    
    hf_TOTAL = CO_Hf + CO2_Hf + H2_Hf + CH4_Hf + H2O_Hf

    CO_VP = 10 ** (CO["A"] - CO["B"] / (T + CO["C"]))
    CO2_VP = 10 ** (CO2["A"] - CO2["B"] / (T + CO2["C"]))
    H2_VP = 10 ** (H2["A"] - H2["B"] / (T + H2["C"]))
    CH4_VP = 10 ** (CH4["A"] - CH4["B"] / (T + CH4["C"]))
    H2O_VP = 10 ** (H2O["A"] - H2O["B"] / (T + H2O["C"]))

    CO_Ki = CO_VP / P
    CO2_Ki = CO2_VP / P
    H2_Ki = H2_VP / P
    CH4_Ki = CH4_VP / P
    H2O_Ki = H2O_VP / P

    CO_Ki_Zi = (CO_Ki - 1) * CO_fraction
    CO2_Ki_Zi = (CO2_Ki - 1) * CO2_fraction
    H2_Ki_Zi = (H2_Ki - 1) * H2_fraction
    CH4_Ki_Zi = (CH4_Ki - 1) * CH4_fraction
    H2O_Ki_Zi = (H2O_Ki - 1) * H2O_fraction

    CO_Ki_VF = 1 + (CO_Ki - 1) * VF_EST
    CO2_Ki_VF = 1 + (CO2_Ki - 1) * VF_EST
    H2_Ki_VF = 1 + (H2_Ki - 1) * VF_EST
    CH4_Ki_VF = 1 + (CH4_Ki - 1) * VF_EST
    H2O_Ki_VF = 1 + (H2O_Ki - 1) * VF_EST

    CO_f = CO_Ki_Zi / CO_Ki_VF
    CO2_f = CO2_Ki_Zi / CO2_Ki_VF
    H2_f = H2_Ki_Zi / H2_Ki_VF
    CH4_f = CH4_Ki_Zi / CH4_Ki_VF
    H2O_f = H2O_Ki_Zi / H2O_Ki_VF

    f_SUM = CO_f + CO2_f + H2_f + CH4_f + H2O_f

    CO_Ki_Ki_Zi = ((CO_Ki - 1) ** 2) * CO_fraction
    CO2_Ki_Ki_Zi = ((CO2_Ki - 1) ** 2) * CO2_fraction
    H2_Ki_Ki_Zi = ((H2_Ki - 1) ** 2) * H2_fraction
    CH4_Ki_Ki_Zi = ((CH4_Ki - 1) ** 2) * CH4_fraction
    H2O_Ki_Ki_Zi = ((H2O_Ki - 1) ** 2) * H2O_fraction

    CO_Ki_VF_Square = CO_Ki_VF ** 2
    CO2_Ki_VF_Square = CO2_Ki_VF ** 2
    H2_Ki_VF_Square = H2_Ki_VF ** 2
    CH4_Ki_VF_Square = CH4_Ki_VF ** 2
    H2O_Ki_VF_Square = H2O_Ki_VF ** 2

    CO_df_dVF = - CO_Ki_Ki_Zi / CO_Ki_VF_Square
    CO2_df_dVF = - CO2_Ki_Ki_Zi / CO2_Ki_VF_Square
    H2_df_dVF = -H2_Ki_Ki_Zi / H2_Ki_VF_Square
    CH4_df_dVF = - CH4_Ki_Ki_Zi / CH4_Ki_VF_Square
    H2O_df_dVF = - H2O_Ki_Ki_Zi / H2O_Ki_VF_Square
    
    CO_bottom = CO_fraction / (1 + (CO_Ki - 1) * VF_EST)
    CO_top = CO_bottom * CO_Ki
    CO2_bottom = CO2_fraction / (1 + (CO2_Ki - 1) * VF_EST)
    CO2_top = CO2_bottom * CO2_Ki
    H2_bottom = H2_fraction / (1 + (H2_Ki - 1) * VF_EST)
    H2_top = H2_bottom * H2_Ki
    CH4_bottom = CH4_fraction / (1 + (CH4_Ki - 1) * VF_EST)
    CH4_top = CH4_bottom * CH4_Ki
    H2O_bottom = H2O_fraction / (1 + (H2O_Ki - 1) * VF_EST)
    H2O_top = H2O_bottom * H2O_Ki
    
    CO_HL = CO_bottom * (CO_HCond + 0)
    CO2_HL = CO2_bottom * (CO2_HCond + 0)
    H2_HL = H2_bottom * (H2_HCond + 0)
    CH4_HL = CH4_bottom * (CH4_HCond + 0)
    
    hl_TOTAL = CO_HL + CO2_HL + H2_HL + CH4_HL
    
    CO_HV = CO_top * CO_CpV * (T_feed - T_ref)
    CO2_HV = CO2_top * CO2_CpV * (T_feed - T_ref)
    H2_HV = H2_top * H2_CpV * (T_feed - T_ref)
    CH4_HV = CH4_top * CH4_CpV * (T_feed - T_ref)
    H2O_HV = H2O_top * H2O_CpV * (T_feed - T_ref)
    
    hV_TOTAL = CO_HV + CO2_HV + H2_HV + CH4_HV + H2O_HV
    
    ek = V * hV_TOTAL + L * hl_TOTAL - Ft * hf_TOTAL
    
    dekdt = V * (CO_top * CO_Cp + CO2_top * CO2_Cp +H2_top * H2_Cp +CH4_top * CH4_Cp + H2O_top * H2O_Cp) + L * (CO_bottom* CO_Cp + CO2_bottom* CO2_Cp + H2_bottom* H2_Cp + CH4_bottom* CH4_Cp + H2O_bottom * H2O_Liq_Cp )
    Tdrum = T - ek / dekdt
    SUM_df_dVF = CO_df_dVF + CO2_df_dVF + H2_df_dVF + CH4_df_dVF + H2O_df_dVF

    VF_squared = VF_EST + f_SUM / -SUM_df_dVF
    print(str(i) + ".Iteration V/F = " + str(round(VF_EST, 5)) + " and f = " + str(round(f_SUM,7)))
    print("Tdrum " + str(round(Tdrum,3)) + " EK = " + str(round(ek,3)))
    VF_EST = VF_squared
```
